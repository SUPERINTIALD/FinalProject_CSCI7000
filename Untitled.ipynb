```python
import os

data_path = './data/test_images'

# Get all files in the directory
files = os.listdir(data_path)

# Sort files by creation time (ctime)
# os.path.join ensures the script finds the file inside the /data/test_images folder
files.sort(key=lambda x: os.path.getctime(os.path.join(data_path, x)))

print(files)
```
